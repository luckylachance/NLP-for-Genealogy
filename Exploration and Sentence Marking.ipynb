#### Loop to read text files in a data folder

```python
import os
import re

path = 'data/'
arr = os.listdir(path)
for filename in arr:
    if filename.endswith('.txt'):
        xpath = os.path.join(path,filename)
        with open(xpath, 'r') as f:
                text = f.read()
                text = text.lower()
                reclean = re.compile('<.*?>')
                clean = re.sub(reclean,'', text)
                cleantext = " ".join(clean.split())
                outFile = open(path + os.path.splitext(filename)[0] +".txt",'w')
                outFile.write(cleantext)
                outFile.close()   
```

```python
import pandas as pd
import nltk
from nltk.tokenize import PunktSentenceTokenizer
from nltk.tokenize import word_tokenize
from nltk.corpus import reuters

tokenizer = PunktSentenceTokenizer()
path = 'data/'
arr = os.listdir(path)
data = []
for filename in arr:
    if filename.endswith('.txt'):
        xpath = os.path.join(path,filename)
        with open(xpath, 'r') as file:
            contents = file.read()
            data.extend(tokenizer.tokenize(contents))
            
```

```python
import nltk
import re 
from nltk.chunk import ne_chunk_sents
from nltk.sem import relextract


def findrelations(cleantext):
    roles = """
    (.*(                   
    husband|
    daughter|
    grandchild).*)|
    Mary Smith|
    (wo)?man|
    man|
    ,\sof\sthe?\s*  # "X, of (the) Y"
    """
    ROLES = re.compile(roles, re.VERBOSE)

    sentences = nltk.sent_tokenize(cleantext)
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
    tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
    chunked_sentences = nltk.ne_chunk_sents(tagged_sentences)


    for document in chunked_sentences:
        print (document)
        for rel in relextract.extract_rels('PER', 'PER', document, corpus='ace', pattern=ROLES):
        
            print (relextract.show_raw_rtuple(rel))

df = findrelations(cleantext)
```

```python
import re 
import string 
import nltk 
import spacy 
import pandas as pd 
import numpy as np 
import math 
from tqdm import tqdm 

from spacy.matcher import Matcher 
from spacy.tokens import Span 
from spacy import displacy 

pd.set_option('display.max_colwidth', 200)
```

```python
# load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# incorporate the text 
text = cleantext

# create a spaCy object 
doc = nlp(text)
```

```python
# print token, dependency, POS tag 
for tok in doc: 
  print(tok.text, "-->",tok.dep_,"-->", tok.pos_)
```

```python
#define the pattern 
pattern1 = [{'POS':'PROPN'},
            {'POS':'PROPN'},
            {'LOWER': 'married'}, 
            {'POS': 'PROPN'},
            {'POS': 'PROPN'}] #proper noun

# Matcher class object 
matcher = Matcher(nlp.vocab) 
matcher.add("matching_1", None, pattern1) 

matches = matcher(doc) 
span = doc[matches[0][1]:matches[0][2]]

print(span.text)
```

```python
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```
